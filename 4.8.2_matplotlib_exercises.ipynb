```python
x = range(-4, 6) 
y = [x**2 - x + 2 for x in x]
plt.plot(x,y, marker='x')
```

```python
# y = x^3
x = (range(1, 100))
y = [n**3 for n in x]

plt.title("$y = x^3$", fontsize=14)

plt.xlabel("x", fontsize=14)
plt.ylabel("$x^3$", rotation=0, fontsize=14)

# plt.yscale("log")
plt.plot(y)
plt.show()
```

```python
x = range(1, 40)
y = [2**n for n in x]

plt.title("$y=2^x$")
plt.xlabel("x")
plt.ylabel("$y=2^x$")

plt.plot(x,y)
plt.show()
```
