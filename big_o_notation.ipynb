```python
import math
```

```python
x = range(1, 20)
y = [x +1 for x in x]

plt.plot(x, y, marker='x')
plt.title('1')
plt.xlabel('Elements')
plt.ylabel('Operation')
```

```python
x = range(1, 20)
y = [math.log(x) for x in x]

plt.plot(x, y, marker='x')
plt.title('log(n)')
plt.xlabel('Elements')
plt.ylabel('Operation')
```

```python
x = range(1, 20)
y = [x for x in x]

plt.plot(x, y, marker='x')
plt.title('n')
plt.xlabel('Elements')
plt.ylabel('Operation')
```

```python
x = range(1, 20)
y = [x*math.log(x) +1 for x in x]

plt.plot(x, y, marker='x')
plt.title('$y = x+1 -- (1)')
plt.xlabel('Elements')
plt.ylabel('Operation')
```

```python
x = range(1, 20)
y = [x**2 for x in x]

plt.plot(x, y, marker='x')
plt.title('n^2')
plt.xlabel('Elements')
plt.ylabel('Operation')
```

```python
x = range(1,20)
y = [math.factorial(n) for n in x]
plt.plot(x,y)
plt.title('n!')
plt.xlabel('Elements')
plt.ylabel('Operation')
```

```python
x = range(1, 20)
y = [2**x for x in x]

plt.plot(x, y, marker='x')
plt.title('2^x')
plt.xlabel('Elements')
plt.ylabel('Operation')
```

```python
x = range(1,20)
y = [x**x for x in x]
plt.plot(x,y)
plt.title('n^n')
plt.xlabel('Elements')
plt.ylabel('Operation')
```

```python
x = range(1, 20)
x_plus_one = [x +1 for x in x]
log_of_x = [math.log(x) for x in x]
x_for_x = [x for x in x]
x_times_log_x = [x*math.log(x) +1 for x in x]
x_squared = [x**2 for x in x]
factorial = [math.factorial(n) for n in x]
two_raised_to_x = [2**x for x in x]
x_raised_to_x = [x**x for x in x]
plt.figure(figsize=(16, 10)) # (width, height)

plt.plot(x, x_plus_one, label='x + 1')
plt.plot(x, log_of_x,  label='log(x)')
plt.plot(x, x_for_x, label='x')
plt.plot(x, x_times_log_x,  label='x*log(x)')
plt.plot(x, x_squared, label='x^2')
plt.plot(x, factorial,  label='x!')
plt.plot(x, two_raised_to_x, label='2^x')
plt.plot(x, x_raised_to_x, label='x^x')

plt.legend(loc='upper right')
plt.title('Lots of lines')
plt.xlabel('$x$')

plt.show()
```
