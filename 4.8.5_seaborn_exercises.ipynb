```python
swiss['is_catholic']=swiss.Catholic>50
```

#### Does whether or not a province is Catholic influence fertility?

```python
df = swiss[['Fertility','is_catholic']]
```

```python
df.head()
```

```python
fertility_influence = df.groupby('is_catholic').Fertility.mean()
```

```python
ax = fertility_influence.plot.bar(x='is_catholic',y='Fertility', rot = 0)
# ax = df.plot.bar(x='lab', y='val', rot=0)
```
